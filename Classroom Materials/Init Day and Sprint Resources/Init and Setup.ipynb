## Ground rules
We have 4 short hours each night to pack in a lot of learning. It’s both a lot of time at the end of a long day, and very little time to accomplish an ambitious objective. As a cohort, everyone -- from the least experienced novice, to the most experienced coach -- relies on each other to support each other’s learning. It sounds corny, but these ground rules are meant to establish trust by setting expectations we can all meet. Please be honest about your ability and willingness to commit to adhering to the rules

#### General
-	Arrive and be ready for standup promptly at 6:00PM. Notify the slack channel if you will be late.
- Be respectful
    -	Presenters get your full attention, no laptops, no phones, no side discussions, unless asked to look something up.

#### Classroom
- Wipe down and clean up surfaces. This is a shared space.
- Use coasters or put drinks with condensation on the floor. The whiteboard tables warp
- Clean up kitchen and food. Use the kitchen trash can for food waste.
- notify the coaching team if you notice anything weird or that might be a safety issue outside.

#### Bringing it
-	Ask every question you can think of (at the appropriate time)
-	Commit to listening thoughtfully.
-	We are all resources for each other. Take responsibility for everyone’s learning objectives.
-	Learning new things puts you in an extremely vulnerable state. Move towards vulnerability, rather than away from it.
-	Communicate honestly and directly. Be critical when necessary, but first make sure you understand (and communicate) what you think is happening
-	Take risks in the classroom. Make comments you wouldn’t normally make. Present things you wouldn’t normally present.
-	Recognize and respect the risks others are taking to learn new things.
-	Learning has to happen in a safe environment. If you sense that the environment is becoming threatening or unsafe in any way, say something. The class safe word is “Banana”
-	Everyone benefits from knowing what you don’t know. Ask questions that you’re embarrassed to ask.
-	If you say you’re going to do something, do it
-	If you think any of these ground rules are being broken, bring it up during sprint retrospective

#### Frustration
Frustration ground rules: I think it’s worth having a separate section on how to deal with frustration because I expect it to be a common feature in the class format. You will get frustrated learning new things. You will get frustrated working with people who are different from you, including your instructor. (I also experience a lot of frustration when learning, so these are ground rules I need to follow as well)

-	Frustration is part of learning. If you’re not frustrated, you’re probably not challenging your understanding
-	Communicate what is frustrating you as clearly as you can
-	Listen to other people’s frustration with an ear towards clearly understanding what they want and what they feel.
-	Don’t assign your frustration to other people
-	Support others’ frustration by asking them one thing that you can do to alleviate the frustration.
