# Skills Backlog / Blueprint

Creating (and periodically revisiting and refining) your __skills backlog__ is one of the most important parts of the skill building process. A well defined skill goal will ensure that you spend your time effectively on something tangible and useful.

#### User Stories
You will create a user story for each sprint to capture the skill you want to build and why. User stories have three parts

1. As a [---------------]
2. I need [---------------]
3. So that [---------------]

#### Developing your Skills Backlog

I recommend taking the following steps to come up with your skills Backlog
1. Go through the blueprint ([pdf](Blueprints.pdf) / [xlsx](Blueprints.xlsx)) and highlight the user story or user stories that seem interesting to you. You might follow a track if you think that would drive your use. Or just go with the ones you're drawn to
2. Take notes about why each seemed interesting to you, particularly if you can think of a specific application in your life or work or future interests.
3. Personalize the user story to include the skill you'd like to build and how you'd like to apply it using the notes and thoughts generated by the process above to guide your development.
4. If you have questions about what something is or what it means, google around a bit. Or ask about it on Slack!

#### Add your Skill Backlog to the Class Trello Board
Create a new __List__ in Trello using your name as the title in the [class Trello board](https://trello.com/b/7586eapN/devleague-big-data-analyst-bootcamp). Create a __card__ for each skill / user story in your skills backlog. Attach a trello green "skill" __label__ to each card and __assign__ it to yourself.
